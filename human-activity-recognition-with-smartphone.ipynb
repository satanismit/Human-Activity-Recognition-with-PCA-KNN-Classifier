```python
# seperate target column for test data 

X_test=df.drop(columns=['Activity'])
y_test=df['Activity']
```

```python
X_test.shape
```

```python
#seperate one target column  for train dataset 
X_train=df.drop(columns=['Activity'])

y_train=df['Activity']
```

```python
y_train.shape
```

**we are simply use knn Classifier**

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier()

knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
from sklearn.metrics import *

print("Accuracy of knn model with all features ",accuracy_score(y_test,y_pred)*100)
```

**WITH PCA**

```python
from sklearn.decomposition import PCA

pca= PCA(n_components=2)
```

```python
X_train_pca=pca.fit_transform(X_train)
X_test_pca=pca.transform(X_test)

X_train_pca.shape
```

```python
X_test_pca.shape
```

```python
X_train.shape
```

Here we see that after applying PCA the shape of the dataset or dimension are reduced upto 2.

```python
%matplotlib inline
```

```python
import plotly.express as px

y_train_pca=y_train.astype(str)

fig= px.scatter(    x=X_train_pca[:,0],
                    y=X_train_pca[:,1],
                    color=y_train_pca,
                    color_discrete_sequence=px.colors.qualitative.G10)
fig.show()
```

In this we see that all categories points but the some of points are hidden in 2d pot so we cant differentiate , let's make us 3D and Flexible 

```python
pca= PCA(n_components=3)

X_train_pca=pca.fit_transform(X_train)
X_test_pca=pca.transform(X_test)
```

```python
y_train_pca=y_train.astype(str)

fig= px.scatter_3d(    x=X_train_pca[:,0],
                    y=X_train_pca[:,1],
                    z=X_train_pca[:,2],
                    color=y_train_pca )
fig.update_layout(
           margin=dict(l=20, r=20, t=20, b=20)
)
                    
fig.show()
```

Here is the best representation in 3D for three principal components...

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(X_train_pca,y_train)
```

```python
y_pred_pca=knn.predict(X_test_pca)
```

```python
print("Accuracy of knn model with Principal components ",accuracy_score(y_test,y_pred_pca))
```

```python
for i in range(20):
    pca= PCA(n_components=i+1) # because of first value is 0

    X_train_pca=pca.fit_transform(X_train)
    X_test_pca=pca.transform(X_test)

    knn=KNeighborsClassifier()
    knn.fit(X_train_pca,y_train)
    
    y_pred_pca=knn.predict(X_test_pca)
    
    print(f"Accuracy of knn model with {i} PC :",accuracy_score(y_test,y_pred_pca))
```

Here we see that if we increase the number of principal components then accuracy of model is also increase.

So if we compare the accuracy of  knn model that include all features and this knn model that uses principal components we get the approximately same accuracy score..

And may be if we calculate with more pc we reach to exact same accuracy 

Let's do this :

```python
for i in range(40):
    pca= PCA(n_components=i+1) # because of first value is 0

    X_train_pca=pca.fit_transform(X_train)
    X_test_pca=pca.transform(X_test)

    knn=KNeighborsClassifier()
    knn.fit(X_train_pca,y_train)
    
    y_pred_pca=knn.predict(X_test_pca)
    
    print(f"Accuracy of knn model with {i} PC :",accuracy_score(y_test,y_pred_pca)*100)
```

```python
# as we have mentioned above we get the same accuracy score -- 98 %
```
